```python
# To find lowest rent in any area of Delhi city only
Delhi = df.groupby('city').get_group('Delhi')
```

```python
df2 = pd.DataFrame(Delhi)
```

```python
df2
```

```python
df2.sort_values(by = 'total_rent')
```

```python
# Area in which flat is available in Delhi city.

list(df2['area'].unique())
```

```python
# To check flats of 2 rooms and 2 bathrooms in Delhi city.

df2.query('rooms == 2 & bathroom == 2')
```

```python
# To check flats of 2 rooms and 2 bathrooms in Vasant Kunj of Delhi city.

df2.query('rooms == 2 & bathroom == 2 & area == "Vasant Kunj"')
```

```python
# To find flat with 2 rooms and furniture having rent under 10000 in Delhi.

df2.query('rooms == 2 & furniture == "furnished" & total_rent < 10000' )
```

```python
# Top 5 flat with 2 rooms and furniture having rent under 10000 in Delhi.

df2.query('rooms == 2 & furniture == "furnished" & total_rent < 10000' ).head()
```

```python
# Top 5 Costliest flat in Delhi.
df2.query('rooms == 2 & furniture == "furnished"' ).tail()
```
